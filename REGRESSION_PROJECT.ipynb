


> **Second Part**

Here I import the Pandas library and the fetch_openml function from scikit-learn to load the Boston Housing Dataset. Then converts the dataset into a Pandas dataframe and prints the first five rows of the dataframe.



```python
import pandas as pd
from sklearn.datasets import fetch_openml

# Load the Boston Housing Dataset
boston = fetch_openml(name='boston', version=1)

# Convert the dataset into a Pandas dataframe
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)

# Print the first 5 rows of the dataset
print(boston_df.head())
```

> **Third part**

Now I add a new column called 'MEDV' to the previously created dataframe 'boston_df' and assigns the target variable values to this column. Then it prints the first five rows of the updated dataframe including the target variable.


```python

# Add the target variable MEDV to the dataframe
boston_df['MEDV'] = boston.target

# Print the first 5 rows of the dataset with target variable
print(boston_df.head())
```



> **Fourth Part**

Here I create a correlation matrix of the Boston Housing dataset using the Pandas .corr() function, rounds the matrix to two decimal places, and then plots it as a heatmap using the Seaborn .heatmap() function with annotations enabled. The heatmap visually displays the correlation between each pair of features in the dataset.



> **Fifth Part**

 Now its time to plot two scatter plots using Matplotlib library. The first plot shows the relationship between the average number of rooms per dwelling (RM) and the median value of owner-occupied homes in $1000s (MEDV), and the second plot shows the relationship between the % lower status of the population (LSTAT) and MEDV. X is a Pandas DataFrame containing the independent variables, and y is a Pandas Series containing the dependent variable.



```python
import matplotlib.pyplot as plt

# Select the features RM and LSTAT as independent variables
X = boston_df[['RM', 'LSTAT']]

# Select MEDV as the dependent variable
y = boston_df['MEDV']

# Plot a scatter plot of RM and MEDV
plt.scatter(X['RM'], y)
plt.xlabel('Average number of rooms per dwelling (RM)')
plt.ylabel('Median value of owner-occupied homes in $1000s (MEDV)')
plt.show()

# Plot a scatter plot of LSTAT and MEDV
plt.scatter(X['LSTAT'], y)
plt.xlabel('% lower status of the population (LSTAT)')
plt.ylabel('Median value of owner-occupied homes in $1000s (MEDV)')
plt.show()
```



> **Sixth Part**

Preparing data for machine learning.

Now I concatenate the 'LSTAT' and 'RM' columns of the Boston Housing dataset, and assigns them to the variable 'X'. The target variable 'MEDV' is assigned to 'y'.

Then I use the train_test_split function from the scikit-learn library to split the data into training and testing sets. The test size is set to 20% of the total data, and a random state of 0 is used for reproducibility. Then print the sizes of the training and test sets.



```python
import numpy as np

# Concatenate LSTAT and RM columns using np.c_
X = np.c_[boston_df['LSTAT'], boston_df['RM']]
y = boston_df['MEDV']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Print the sizes of the training and test sets
print("Training set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
```

```python
from sklearn.linear_model import LinearRegression

# create linear regression object
model = LinearRegression()

# fit the model on the training data
model.fit(X_train, y_train)

# make predictions on both training and test data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# evaluate model performance on training set
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
r2_train = r2_score(y_train, y_train_pred)

print("Model performance on training set:")
print("RMSE: {:.2f}".format(rmse_train))
print("R^2 score: {:.2f}".format(r2_train))

# evaluate model performance on testing set
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2_test = r2_score(y_test, y_test_pred)

print("\nModel performance on testing set:")
print("RMSE: {:.2f}".format(rmse_test))
print("R^2 score: {:.2f}".format(r2_test))
```
